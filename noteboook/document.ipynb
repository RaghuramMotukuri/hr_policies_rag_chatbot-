```python
### textloder to read the text files and create document objects
from langchain_community.document_loaders import TextLoader

loader = TextLoader (
    "../data/text_files/python_into.txt",encoding="utf-8"
)
document = loader.load()
print(document)
```

```python
### directory Loader
from langchain_community.document_loaders import DirectoryLoader


dir_loader = DirectoryLoader(
    "../data/text_files",
    glob="**/*txt",## pattern to match file 
    loader_cls= TextLoader, ## loader class to use 
    loader_kwargs={"encoding":"utf-8"},
    show_progress=False
)

documents = dir_loader.load()
print(type(documents))

documents
```

```python
import site, os, shutil, glob

site_packages = site.getsitepackages()[0]
for pattern in ["jsonpatch*"]:
    for path in glob.glob(os.path.join(site_packages, pattern)):
        print("Removing:", path)
        if os.path.isdir(path):
            shutil.rmtree(path)
        else:
            os.remove(path)
```
